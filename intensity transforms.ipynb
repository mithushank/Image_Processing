```python
#identity transformation
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
transform = np.arange(0,256).astype("uint8")
fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlim(0,255)
ax.set_ylim(0,255)
plt.show()

im_o=cv.imread("bw1.jpeg", cv.IMREAD_GRAYSCALE)

im_t = cv.LUT(im_o, transform)
cv.imshow("image_transformed", im_t)
cv.waitKey(0)
cv.destroyAllWindows()


```

```python
#inverse 
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
transform = np.arange(255,-1,-1).astype("uint8")
fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlim(0,255)
ax.set_ylim(0,255)
plt.show()

im_o=cv.imread("images/bw_girl1.jpeg", cv.IMREAD_GRAYSCALE)
cv.imshow("image_original", im_o)
cv.waitKey(0)
cv.destroyAllWindows()

im_t = cv.LUT(im_o, transform)
cv.imshow("image_transformed", im_t)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#intensity windowing 
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

c = np.array( [ ( 100,50) , ( 150,200) ] )
t1 = np.linspace(0, c[0,1], c[0,0]+1-0).astype("uint8")

t2 = np.linspace(c[0,1]+1, c[1,1],c[1,0]-c[0,0]).astype("uint8")

t3 = np.linspace(c[1,1], 256,255-c[1,0]).astype("uint8")

transform = np.concatenate((t1,t2),axis=0).astype("uint8")
transform = np.concatenate((transform, t3),axis=0).astype("uint8")

fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlim(0,255)
ax.set_ylim(0,255)
plt.show()

im_o=cv.imread("images/bw_girl1.jpeg", cv.IMREAD_GRAYSCALE)
#cv.namedWindow("image", cv.WINDOW_AUTOSIZE)
cv.imshow("image_original", im_o)
#cv.waitKey(0)
#cv.destroyAllWindows()

im_t = cv.LUT(im_o, transform)
cv.imshow("image", im_t)
cv.waitKey(0)
cv.destroyAllWindows()
```
