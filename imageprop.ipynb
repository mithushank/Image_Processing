```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
im = cv.imread("bw1.jpeg", cv.IMREAD_COLOR)
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)
fig , ax = plt.subplots()
ax.imshow(im)
ax.set_title("Image")
plt.show()
print(im.shape, im.size, im.dtype)
```

```python
#increasing the brightness
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
im = cv.imread("bw1.jpeg", cv.IMREAD_GRAYSCALE)
imb = im + 100
imc = cv.add(im,100)
fig , ax = plt.subplots(1,3)
ax[0].imshow(im, cmap="gray")

ax[1].imshow(imb, cmap="gray")

ax[2].imshow(imc, cmap="gray")


plt.show()
```

```python
#Zeroing out the green and blue planes
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt


im = cv.imread("images/c1.jpg", cv.IMREAD_ANYCOLOR)
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)

plt.imshow(im)
plt.title("Image")
plt.xticks([]), plt.yticks([])
plt.show()

im[:,:,2:3]=0
plt.imshow(im)
plt.title("After Zeroing Planes")
plt.xticks([]), plt.yticks([])
plt.show()
```
