```python
#averaging kernel
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = cv.imread("images/vk1.jpg", cv.IMREAD_REDUCED_GRAYSCALE_2)
kernel = np.ones((4,4),np.float32)/9

imc = cv.filter2D(im,-1, kernel)

f, a =plt.subplots(1,2, sharex='all', sharey='all', figsize=(18,18))
a[0].imshow(im, cmap='gray')

a[1].imshow(imc, cmap= "gray")

plt.show()
```

```python
#sober vertical kernel
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = cv.imread("images/vk1.jpg", cv.IMREAD_REDUCED_GRAYSCALE_2)
kernel = np.array([(-1,-2,-1),(0,0,0),(1,2,1)], dtype='float')
imc = cv.filter2D(im,-1, kernel)

f, a =plt.subplots(1,2, sharex='all', sharey='all', figsize=(18,18))
a[0].imshow(im, cmap='gray')

a[1].imshow(imc, cmap= "gray")

plt.show()
```
