```python
#plotting gamma graphs
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

fig, ax =plt.subplots()
gamma = [0.2, 0.5, 1, 2, 5]
for g in gamma:
    table = np.array([(i/255.0)**(g)*255.0 for i in np.arange(0,256)])
    ax.plot(table, label = '$\gamma = {g}$'.format(g=g))
ax.legend(loc ='best')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
plt.show()
```

```python
#gamma correction
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im1 = cv.imread("images/c2.jpeg", cv.IMREAD_COLOR)
gamma = 50
table = np.array([(i/255.0)**(g)*255.0 for i in np.arange(0,256)]).astype('uint8')
im_gamma1 = cv.LUT(im1, table)
im_o =cv.cvtColor(im1, cv.COLOR_BGR2RGB)
im_gamma =cv.cvtColor(im_gamma1, cv.COLOR_BGR2RGB)
f, a =plt.subplots(3,2)
a[0,0].imshow(im_o)
a[0,1].imshow(im_gamma1)
color = ('b','g','r')
for i,c in enumerate(color):
    hist1 = cv.calcHist([im_o], [i], None, [256], [0,256])
    a[1,0].plot(hist1, color = c)
    hist2 = cv.calcHist([im_gamma1], [i], None, [256], [0,256])
    a[1,1].plot(hist2, color = c)

a[2,0].plot(table)
a[2,0].set_xlim(0,255)
a[2,0].set_ylim(0,255)
a[2,0].set_aspect('equal')
```
